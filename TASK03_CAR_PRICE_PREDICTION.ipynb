```python
categorical_features = ['fueltype', 'aspiration','doornumber', 'carbody', 'drivewheel', 'enginelocation',  'enginetype', 'cylindernumber', 'fuelsystem']
numerical_features = ['wheelbase',	'carlength',	'carwidth',	'carheight',	'curbweight',	'enginesize',	'boreratio	stroke',	'compressionratio',	'horsepower',	'peakrpm',	'citympg', 'highwaympg'	,'price']
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
for feature in categorical_features:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=feature, y='price', data=data)
    plt.title(f'{feature} vs Price')
    plt.show()
```

```python
from scipy.stats import f_oneway
```

```python
for feature in categorical_features:
    categories = data[feature].unique()
    groups = [data[data[feature] == category]['price'] for category in categories]
    f_statistic, p_value = f_oneway(*groups)
    print(f"{feature} - F-statistic: {f_statistic:.4f}, p-value: {p_value:.4f}")
```

```python
X = data.drop(columns=['price'])
y = data['price']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Evaluate the model on the testing data

y_pred = rf_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```
